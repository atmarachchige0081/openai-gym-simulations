```python
## Frozen Lake Example
import gymnasium as gym

env = gym.make("FrozenLake-v1", map_name="4x4", is_slippery=True, render_mode="human")

max_steps = 100  

num_episodes = 5

for episode in range(num_episodes):
    observation, info = env.reset()
    done = False
    total_reward = 0.0
    timestep = 0
    
    while not done and timestep < max_steps:

        env.render()

        action = env.action_space.sample()

        observation, reward, done, truncated, info = env.step(action)

        total_reward += reward

        timestep += 1

        print(f"Step: {timestep}")
        print(f"Observation: {observation}")
        print(f"Action: {action}")
        print(f"Reward: {reward}")
        print(f"Done: {done}")
        print(f"Truncated: {truncated}")
        print("-----------------------------")

        if done or truncated:
            break
    
    print(f"Episode {episode + 1} ended with total reward: {total_reward}")

env.close()
```

```python
import gymnasium as gym
import numpy as np

env = gym.make("MountainCarContinuous-v0", render_mode="human")

num_episodes = 1

for episode in range(num_episodes):
    observation, info = env.reset()
    done = False
    total_reward = 0.0
    timestep = 0
    
    while not done:
        
        env.render()

        # Action strategy: Encourage exploration in both directions
        if observation[1] < 0:  # If velocity is negative (moving left)
            action = np.array([0.5])  # Push moderately to the right
        else:
            action = np.array([1.0])  # Push strongly to the right

        observation, reward, done, truncated, info = env.step(action)

        # Improved reward function
        position, velocity = observation
        goal_position = env.goal_position
        
        # Reward for moving towards the goal
        reward = (position - (-0.6)) / (goal_position - (-0.6))
        
        # Penalize large actions
        reward -= 0.1 * np.abs(action[0])
        
        # Additional small penalty for moving left
        if velocity < 0:
            reward -= 0.05

        total_reward += reward
        timestep += 1

        print(f"Step: {timestep}")
        print(f"Observation: {observation}")
        print(f"Action: {action}")
        print(f"Reward: {reward}")
        print(f"Done: {done}")
        print(f"Truncated: {truncated}")
        print("-----------------------------")

        if done or truncated:
            break
    
    print(f"Episode {episode + 1} ended with total reward: {total_reward}")

env.close()
```

```python
import gymnasium as gym

# Initialize the CartPole environment
env = gym.make('CartPole-v1', render_mode='human')

# Reset the environment to start a new episode
state = env.reset()

# Set a maximum number of steps per episode
max_steps = 500

# Run the environment for a fixed number of steps
for step in range(max_steps):
    # Render the current state of the environment
    env.render()
    
    # Randomly select an action (0 or 1) to apply
    action = env.action_space.sample()
    
    # Apply the selected action and get the new state, reward, and done flag
    state, reward, done, _, _ = env.step(action)
    
    # Print the current step, action, and reward
    print(f"Step: {step}, Action: {action}, Reward: {reward}")
    
    # Check if the episode is done (pole fell over or max steps exceeded)
    if done:
        print("Episode finished")
        break

# Close the environment
env.close()
```

```python
import gymnasium as gym
import numpy as np

env = gym.make("CarRacing-v2", render_mode="human")

observation, info = env.reset()

done = False
total_reward = 0.0


def driving_policy():#+
    """#+
    This function generates a driving policy for the CarRacing-v2 environment.#+
    The policy randomly selects steering, gas, and brake values within specified ranges.#+
#+
    Parameters:#+
    None#+
#+
    Returns:#+
    np.array([steering, gas, brake]): A numpy array containing the steering, gas, and brake values.#+
    """#+
    steering = np.random.uniform(-0.000001, 0.00001)  # Randomly select a steering value between -0.0 and 0.0#+
    gas = 0.9  # Set the gas value to 0.9#+
    brake = 0.0  # Set the brake value to 0.0#+
    return np.array([steering, gas, brake])  # Return the steering, gas, and brake values as a numpy array#+

while not done:

    action = driving_policy()
    
    observation, reward, done, truncated, info = env.step(action)
    
    total_reward += reward
    
    if info.get('off_track', False):
        print("Car went off the track. Resetting the environment.")
        observation, info = env.reset()
        continue 
    
    env.render()

env.close()

print(f"Total reward: {total_reward}")
```
